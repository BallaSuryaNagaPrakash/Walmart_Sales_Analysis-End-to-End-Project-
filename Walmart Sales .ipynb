# Visulization 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Walmart_Sales.csv")

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df
```

# 1. Univarient Analysis 

# 1. Weekly Sales Distribution

```python
sns.histplot(df['Weekly_Sales'], bins=50)
plt.title("Distribution of Weekly Sales")
plt.show()
```

# 2. Temperature Distribution

```python
sns.boxplot(x=df['Temperature'])
plt.title("Temperature Distribution")
plt.show()
```

# 3.Holiday Frequency

```python
df['Holiday_Flag'].value_counts().plot(kind='bar')
plt.title("Holiday vs Non-Holiday Weeks")
plt.show()
```

# 2. Bivariate Analysis (Two Variables)

# 1. Sales vs Holiday

```python
sns.boxplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
plt.title("Sales during Holiday vs Non-Holiday")
plt.show()
```

# 2. Temperature vs Sales

```python
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.title("Temperature vs Weekly Sales")
plt.show()
```

# 3. Store Performance

```python
store_sales = df.groupby('Store')['Weekly_Sales'].sum()

store_sales.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top Performing Stores")
plt.show()
```

# 3. Multivariate Analysis (Multiple Variables)

# 1. Correlation Analysis

```python
corr = df[['Weekly_Sales','Temperature','Fuel_Price','CPI','Unemployment']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

# 2. Monthly Sales Trend

```python
monthly_sales = df.groupby('Month')['Weekly_Sales'].sum()

monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.show()
```

# 3.Sales by Store and Holiday

```python
pivot = pd.pivot_table(df, values='Weekly_Sales', index='Store', columns='Holiday_Flag')

sns.heatmap(pivot)
plt.title("Store Sales during Holidays")
plt.show()
```

# 4.Advanced Industry Analysis

# 1.Sales Trend Over Time

```python
df.groupby('Date')['Weekly_Sales'].sum().plot(figsize=(10,5))
plt.title("Overall Sales Trend")
plt.show()
```

# 2.Store Contribution to Total Revenue

```python
store_share = df.groupby('Store')['Weekly_Sales'].sum()

store_share.plot(kind='pie')
plt.title("Store Contribution to Total Sales")
plt.show()
```
